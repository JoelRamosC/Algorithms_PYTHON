Removendo dados duplicados

```python
dados = {
    'coluna 1':[1,1,2,2,3,3,3],
    'coluna 2':['a','a','b','b','c','c','c'],
    'coluna 3':['A','A','B','B','C','C','C']
}
df = DataFrame(dados)
df
```

```python
#identifica as linhas com valores repetidos
df.duplicated()
```

```python
#limina linhas com repetição
df.drop_duplicates()
```

```python
#atualiza o df e elimina as linhas com repetição definitivo
df.drop_duplicates(inplace=True)
df
```

```python
#considera os repetidos apenas da coluna 3
df.drop_duplicates(["coluna 3"])
```

Concatenando Dados

```python
df = pd.DataFrame(np.arange(36).reshape((6,6)))
df
```

```python
df2 = pd.DataFrame(np.arange(15).reshape(5,3))
df2
```

```python
#concatenar dataframes na vertical, somando linhas
pd.concat([df,df2])
```

```python
#concatena na lateral, somando colunas
pd.concat([df,df2], axis =1)
```
